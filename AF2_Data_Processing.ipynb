**PART 1 DATA LOADING AND EXPLORATION**

```python
# 1. Load the dataset using Pandas.
import pandas as pd
df = pd.read_csv("diabetes.csv")
  # 2. Display the first few rows of the dataset.
df.head()
  # 3. Identify null or inconsistent values (zero values in columns such as Glucose, BloodPressure, BMI, etc.).
(df == 0).sum()
  # 4. Generate descriptive statistics using .describe().
df.describe()
```

**PART 2 DATA CLEANING AND NORMALIZATION**

```python
# 1. Replace zero values in attribute columns with the median of each column.
df_clean = df.copy()
columns_to_fix = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
for col in columns_to_fix:
    median_value = df_clean[col].median()
    df_clean[col] = df_clean[col].replace(0, median_value)

# 2. Normalize numerical data using Min-Max Scaling.
df_normalized = df_clean.copy()
for col in df_normalized.columns:
    if col != "Outcome":
        df_normalized[col] = (
            df_normalized[col] - df_normalized[col].min()
        ) / (
            df_normalized[col].max() - df_normalized[col].min()
        )
```

**PART 3 DATA VISUALIZATION**
*STEP 1 (HISTOGRAM - BEFORE)*




```python
 # 1. Histogram of glucose levels BEFORE normalization.
import matplotlib.pyplot as plt
plt.hist(df_clean["Glucose"], bins=20)
plt.title("Glucose Distribution (Before Normalization)")
plt.xlabel("Glucose")
plt.ylabel("Frequency")
plt.show()
```

*STEP 1 (HISTOGRAM - AFTER)*

```python
  # 1. Histogram of glucose levels AFTER normalization.
plt.hist(df_normalized["Glucose"], bins=20)
plt.title("Glucose Distribution (After Normalization)")
plt.xlabel("Normalized Glucose")
plt.ylabel("Frequency")
plt.show()

```

1. What impact do missing values have on data analysis?

Missing values, which appear in data sets as NaN or Unknown, have a significant impact during the process for data analysis in machine learning, since they can create biases, break algorithms and reduce accuracy as a whole.

2. How can we detect and handle outliers in biomedical datasets?

Outliers are values that differ significally from the ones we found on our data set. For example, in our data set one outlier would be the value 0 found across the descriptive statistics.

Outlier values can affect the analysis of our data, reducing the accuracy of our results. We can handle this type of data by applying visual techniques (Box Plots, Scatter Points, Error Bars...), or by using statistical methods.

Once we identified the outlier, we can either trim, remove or modify their values to better fit our data set.

3. What are some common causes of missing or incorrect data in biomedical studies?

Incorrect data can occur for a variety of reasons. Errors can occur during the measurement, analysis and presentation of data, and they can lead to inaccurate results in biomedical studies. Furthermore, faulty instruments or poorly designed experimentation can lead to mistakes during the first stages of data collection, causing both missing and incorrect data.

4. How does normalizing data improve the performance of Machine Learning models?

Normalizing data improves the performance because it makes all the variables comparable by being on the same scale. This is relevant when the variables are measured in different units and ranges, which creates a bias in the learning process. When the data is normalized, all the variables are rescaled to be in range between 0 and 1, and this has as result a faster and more  reliable training.

5. Why is visualizing data an important step before applying Machine Learning models?

It is important because we need to ensure that what we are giving to the model is reliable. Using graphs like the histograms in our case, helps us to spot issues such as impossible values or missing data, thus making us to fix the data before giving it to the model. Visualizing the data is important because it gives us an idea on how to preprocess the data.

6. Why is preprocessing critical when working with medical datasets?

Preprocessing is made in order to prepare the data for training a model. When working with medical datasets, the data represents information from real patients, and if it is not handled correctly, the model may learn from that data that is not possible or is not real. When the model learns something that  physiologically is not, but numerically makes sense, it can lead to classify a patient as healthy when they could be at risk. The datasets has to be preprocessed carefully before giving it to the model so what the model learns is correct and makes sense,and thus making the outcome reliable.

7. What ethical concerns should be considered when handling patient data?

Patient confidentiallity is one of the main concerns when handling data from healthcare centers, since any security breach or data mishandling can lead to the exposure of sensitive information, such as medical records and personal identifiers, to unauthorized parties. This threatens both with the confidentiality of the patient, as well as causing a possible risk to their wellbeing and protection.
